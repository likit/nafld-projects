```python
import pandas as pd
```

```python
services = pd.read_csv('/Users/likit/Downloads/NAFLD-data/Services2011.csv')
```

```python
services = services.set_index('cmsCode')
```

```python
physical_exam_df = pd.read_csv('/Users/likit/Downloads/NAFLD-data/Physical2011.csv')
```

```python
physical_exam_df = physical_exam_df.set_index('ServiceNo')
```

```python
tg = pd.read_csv('/Users/likit/Downloads/NAFLD-data/TG2011.csv')
```

```python
tg = tg.set_index('ServiceNo')
```

```python
hdl = pd.read_csv('/Users/likit/Downloads/NAFLD-data/HDLC2011.csv')
```

```python
hdl = hdl.set_index('ServiceNo')
```

```python
ua = pd.read_csv('/Users/likit/Downloads/NAFLD-data/UA2011.csv')
ua = ua.set_index('ServiceNo')
```

```python
customer_df = pd.read_csv('/Users/likit/Downloads/NAFLD-data/EmployeeAll.csv')
```

```python
customer_df = customer_df.set_index('cmsCode')
```

```python
customer_df
```

```python
dataset = services.join(customer_df, on='cmsCode')
```

```python
len(dataset)
```

```python
dataset = dataset.set_index('ServiceNo')
```

```python
dataset = dataset.join(physical_exam_df).join(hdl, rsuffix='_hdl').join(tg, rsuffix='_tg').join(ua)
```

```python
del dataset['ServiceDate_hdl']
del dataset['ServiceDate_tg']
del dataset['TCode_tg']
```
